```python
import subprocess

def extract_audio_ffmpeg(input_video, output_audio, start_time):
    command = [
        "ffmpeg", 
        "-ss", start_time, 
        "-i", input_video, 
        "-vn", 
        "-acodec", "libmp3lame",
        "-q:a", "2",
        output_audio
    ]
    try:
        subprocess.run(command, check=True)
        print(f"ready {output_audio}")
    except subprocess.CalledProcessError as e:
        print(f"error fFmpeg: {e}")

extract_audio_ffmpeg("/mnt/c/video/video.mp4", "/mnt/c/video/output_audio.mp3", "00:02:15")
```
